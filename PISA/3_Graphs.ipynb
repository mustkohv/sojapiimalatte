```python
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv("clean_PISA.csv")
country = ["Finland","Estonia","Latvia","Denmark","Korea, Rep.","Australia"]
data["Indicator Name"].unique()

to_cluster = data[data["Indicator Name"].isin(['PISA: Mean performance on the mathematics scale','PISA: Mean performance on the reading scale','PISA: Mean performance on the science scale'])]
frames = []

r = []    
for row in to_cluster["Country Name"].unique():
    df = to_cluster[to_cluster["Country Name"] == row][["2015"]]
    try:
        df.index =["Maths", "Reading", "Science"]
        df_out = df.stack()
        df_out.index = df_out.index.map('{0[1]}_{0[0]}'.format)
        df = df_out.to_frame().T
        df["Country"] = row
        r.append(df)  
    except: #US
        pass
y = pd.concat(r)
y = y[y["Country"].isin(country)]
y.index=y["Country"]


import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.25
 
# set height of bar
bars1 = list(y["2015_Maths"])
bars2 = list(y["2015_Reading"])
bars3 = list(y["2015_Science"])

 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
colors = ["#2A363B","#E84A5F","#FF847C","#FECEAB","#99B898","#355C7D"]

# Make the plot
plt.bar(r1, bars1, color=colors[5], width=barWidth, edgecolor='white', label='Mathematicss')
plt.bar(r2, bars2, color=colors[4], width=barWidth, edgecolor='white', label='Reading')
plt.bar(r3, bars3, color=colors[1], width=barWidth, edgecolor='white', label='Science')
  
# Add xticks on the middle of the group bars
#plt.xlabel('group', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))],sorted(["Finland","Estonia","Latvia","Denmark","Korea","Australia"]))
 
# Create legend & Show graphic
plt.yticks([r for r in [0,100,200,300,400,500,600,700]],[0,100,200,300,400,500,600,""])
plt.legend()
plt.title("Mean scores")
plt.savefig("aaa.png")
plt.show()
```
