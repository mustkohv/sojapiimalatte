```python
import pandas as pd
data = pd.read_csv("EdStatsData.csv")
```

```python
new = data["Indicator Name"].unique()
needed = []
for row in new:
    if "PISA" in row:
        needed.append(row)
```

```python
data_f= data[data["Indicator Name"].isin(needed)]
```

```python
data_f = data_f[["Country Name","Country Code","Indicator Name","Indicator Code","2006","2009","2012","2015"]]
```

At least 1 values - 77 countries
At least 2 values - 71 countries
At least 3 values - 62 countries
At least 4 values - 56 countries
At least 5 values - 45 countries
At least 6 values - 33 countries

```python
#Drop rows with Na
data_f = data_f.dropna(thresh = 8) 

#Select those indicators, where Estonia is represented
e = data_f[data_f["Country Name"] == "Estonia"]
need_those = e["Indicator Name"].unique()
clean = data_f[data_f["Indicator Name"].isin(need_those)]
clean
```

```python
#Saving to csv
clean.to_csv("clean_PISA.csv")
```
